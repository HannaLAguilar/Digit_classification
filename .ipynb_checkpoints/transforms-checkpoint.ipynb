```python
import numpy as np
import matplotlib.pyplot as plt

import cv2
from PIL import Image

import torch
from torchvision import transforms
```

## load in cv2

```python
img = cv2.imread('cat.jpg')[:, :, ::-1]
print(type(img))
print(img.shape)
plt.imshow(img);
```

## load in PIL

```python
img = Image.open('cat.jpg')
print(img)
print(img.size)
plt.imshow(img);
```

## To PIL to np.array

```python
img = Image.open('cat.jpg')
img = np.asanyarray(img)
print(type(img))
print(img.shape)

print('max:', np.max(img), 'min:', np.min(img))
```

## To np.array to PIL

```python
img = cv2.imread('cat.jpg')[:, :, ::-1]
img = Image.fromarray(np.uint8(img))
print(img)
```

## RandomHorizontalFlip

## Normalize

```python
mean = [0.485, 0.456, 0.406]
std = [0.229, 0.224, 0.225]

transform_norm = transforms.Compose([transforms.ToTensor(), transforms.Normalize(mean, std)])

img_norm = transform_norm(img)
img_norm = img_norm.numpy()
print('before transpose img_norm.shape:', img_norm.shape)
img_norm = np.transpose(img_norm, (1, 2, 0))
print('after transpose img_norm.shape:', img_norm.shape)
plt.imshow(img_norm);
```

## Grayscale

```python
transform_gray = transforms.Grayscale()
img_gray = transform_gray(img)
img_gray_np = np.asanyarray(img_gray)
print(img_gray_np.shape)
plt.imshow(img_gray, cmap='gray');
```

## Compose

```python
transform = transforms.Compose([
        transforms.RandomResizedCrop(224),
        transforms.RandomHorizontalFlip(),
        transforms.ToTensor(),
        transforms.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225])
    ])
    
img_comp = transform(img)
img_comp = img_comp.numpy()
img_comp = np.transpose(img_comp, (1, 2, 0))
plt.imshow(img_comp);
```

```python
transform = transforms.Compose([
        transforms.RandomResizedCrop(224),
        transforms.Grayscale(),
        transforms.RandomHorizontalFlip(),
        transforms.ToTensor(),
    ])
    
img_comp = transform(img).squeeze(0)
img_comp = img_comp.numpy()
plt.imshow(img_comp, cmap='gray');
```
